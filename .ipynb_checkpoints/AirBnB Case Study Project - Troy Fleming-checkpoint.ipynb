```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import seaborn as sns

air_bnb = pd.read_csv('../projects/AB_NYC_2019.csv')
air_bnb_dict = air_bnb.to_dict()
air_bnb_DF = pd.DataFrame.from_dict(air_bnb_dict)
print(air_bnb_DF.keys())
```

```python
# Which hosts are the busiest and why?

host_df = pd.DataFrame.from_dict(air_bnb)
host_count = host_df['host_name'].value_counts().head(10)
print(host_count)
```

```python
# How many neighborhood groups are available and which shows up the most?

nbg_df = pd.DataFrame.from_dict(air_bnb)
nbg_count = nbg_df['neighbourhood_group'].value_counts()
nbg_count

nbg_bar = nbg_count.plot(kind='bar', rot=0, ylabel='Number of Occurrances', xlabel = 'Neighborhood Group', legend = False)

plt.show()

# Unique counts of each neighborhood group:

# Manhatten: 21661
# Brooklyn: 20104
# Queens: 5666
# Bronx: 1091
# Staten Island: 373

# There are 5 neighborhood groups in total, and Manhatten shows up the most.
```

```python
# Are private rooms the most popular in manhattan?

options = ['Manhattan']
mt_rooms_count = air_bnb_DF[air_bnb_DF['neighbourhood_group'].isin(options)]

rooms_df = mt_rooms_count.groupby(['neighbourhood_group','room_type']).size()
print(rooms_df)

mt_rooms_bar = rooms_df.plot(kind='bar', figsize=(10,3), ylabel='Room Type Count', xlabel = 'Room Type', legend = False)

plt.show()

# Unique counts of each room type in Manhattan:

# Entire home/apt: 13199
# Private room: 7982
# Shared room: 480

# Private rooms are NOT the most popular choice in Manhattan. Entire homes/apts are.
```

```python
# Which neighorhood group has the highest average price?

nbg_price_df = pd.DataFrame.from_dict({
    'neighbourhood_group': air_bnb.neighbourhood_group,
    'price': air_bnb.price
})

nbg_price_grouped = nbg_price_df.groupby('neighbourhood_group').mean()
nbg_price_grouped

nbg_price_box = nbg_price_grouped.plot(kind='bar', rot=0, ylabel='Average Price', xlabel = 'Neighborhood Group', legend = False)

print(nbg_price_grouped)
plt.show()

# Mean/average price per rental by neighborhood group:

# Bronx: $87.50
# Brooklyn: $124.48
# Manhattan: $196.88
# Queens: $99.50
# Staten Island: $114.81

# Manhattan has the highest average price of all neighborhood groups at $197.
```

```python
# Which neighbor hood group has the highest total price?

nbg_total_price_df = pd.DataFrame.from_dict({
    'neighbourhood_group': air_bnb.neighbourhood_group,
    'total_price': air_bnb.price
})

nbg_total_price_grouped = nbg_total_price_df.groupby('neighbourhood_group').sum()
nbg_total_price_grouped

nbg_total_price_box = nbg_total_price_grouped.plot(kind='bar', rot=0, ylabel='Average Price', xlabel = 'Neighborhood Group', legend = False)

print(nbg_total_price_grouped)
plt.show()

# Total price of each neighborhood group:

# Bronx: $95,459
# Brooklyn: $2,500,600
# Manhattan: $4,264,527
# Queens: $563,867
# Staten Island: $42,825

# Manhattan has the highest total price of all neighborhood groups at $4,265,527.
```
